```python
from scipy.stats import chi2_contingency, ttest_ind

## Hypothesis I: Cancellation rates
# Create a contingency table for cancellation and lead time (> 180 days)
contingency_table = pd.crosstab(df["is_canceled"], df["lead_time"] > 180)

# Perform chi-square test
chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
print("Hypothesis 1: Chi-square statistic:", chi2_stat, "p-value:", p_value)

##Hypothesis II: ADR
# Create two groups for weekday and weekend bookings
weekday_adr = df[df["stays_in_week_nights"] > 0]["adr"]
weekend_adr = df[df["stays_in_weekend_nights"] > 0]["adr"]

# Perform t-test
t_stat, p_value_ttest = ttest_ind(weekday_adr, weekend_adr, equal_var=False)
print("Hypothesis 2: t-statistic:", t_stat, "p-value:", p_value_ttest)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load the dataset (assuming df is already loaded)

# Step 1: Define X (features) and y (target)
X = df.drop(columns=['is_canceled'])  # All features except 'is_canceled'
y = df['is_canceled']  # Target variable is 'is_canceled'


# Categorical and numerical columns
categorical_cols = ['hotel', 'arrival_date_year', 'arrival_date_month', 'meal', 'market_segment', 
                    'distribution_channel', 'reserved_room_type', 'assigned_room_type', 'deposit_type', 
                    'customer_type']

numerical_cols = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 
                  'babies', 'previous_cancellations', 'previous_bookings_not_canceled', 'booking_changes', 
                  'days_in_waiting_list', 'adr', 'required_car_parking_spaces', 'total_of_special_requests']

# Step 3: Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Step 4: Build Pipelines for Logistic Regression and Random Forest
# Define preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

# Logistic Regression Model with increased max_iter
logistic_model = Pipeline(steps=[('preprocessor', preprocessor), 
                                  ('classifier', LogisticRegression(max_iter=1000))])


# Random Forest Model
random_forest_model = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', RandomForestClassifier())])

# Step 5: Train the models
logistic_model.fit(X_train, y_train)
random_forest_model.fit(X_train, y_train)

# Predictions
y_pred_logistic = logistic_model.predict(X_test)
y_pred_rf = random_forest_model.predict(X_test)

# Step 6: Model Evaluation
# Logistic Regression Metrics
logistic_accuracy = accuracy_score(y_test, y_pred_logistic)
logistic_precision = precision_score(y_test, y_pred_logistic)
logistic_recall = recall_score(y_test, y_pred_logistic)
logistic_f1 = f1_score(y_test, y_pred_logistic)

print("Logistic Regression Metrics:")
print(f"Accuracy: {logistic_accuracy}")
print(f"Precision: {logistic_precision}")
print(f"Recall: {logistic_recall}")
print(f"F1-Score: {logistic_f1}")

# Random Forest Metrics
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_precision = precision_score(y_test, y_pred_rf)
rf_recall = recall_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf)

print("\nRandom Forest Metrics:")
print(f"Accuracy: {rf_accuracy}")
print(f"Precision: {rf_precision}")
print(f"Recall: {rf_recall}")
print(f"F1-Score: {rf_f1}")
```
